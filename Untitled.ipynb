```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv('/content/International_Report_Departures.csv')
data1.head()
```

```python
data2=pd.read_csv('/content/International_Report_Passengers.csv')
data2.head()
```

```python
data=pd.concat([data1,data2],axis=0)
data.index=data.data_dte
data.drop('data_dte',axis=1,inplace=True)
data.head()
```

```python
data.dropna(inplace=True)
```

```python
sns.countplot(x='type',data=data)
plt.title('Distrubtion Of Type',fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
data.carrier.value_counts()
```

```python
data.carriergroup.value_counts()
```

```python
sns.countplot(x='carriergroup',data=data)
plt.title('Distrubtion Of CarrierGroup',fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
box=data.loc[:,['usg_wac','fg_wac','Charter','Total']]
box
```

```python
plt.figure(figsize=(10,15))
x=1
for i,j in zip(['usg_wac','fg_wac','Charter','Total'],['red','green','black','yellow']):
  plt.subplot(2,2,x)
  sns.boxplot(x=i,data=data,color=j)
  plt.ylabel('Value',fontsize=20)
  plt.title(f'BoxPlot For {i}',fontsize=20)
  plt.xticks(rotation=45)
  x+=1
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Total', data=data, color='deepskyblue')
plt.xlabel('Year')
plt.ylabel('Total Flights')
plt.title('Total Flights by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Total',hue='carriergroup',data=data, color='deepskyblue')
plt.xlabel('Year')
plt.ylabel('Total Flights')
plt.title('Total Flights by Year')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Total',hue='type',data=data, color='deepskyblue')
plt.xlabel('Year')
plt.ylabel('Total Flights')
plt.title('Total Flights by Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Scheduled', data=data, label='Scheduled Flights', color='deepskyblue')
sns.lineplot(x='Year', y='Charter', data=data, label='Chartered Flights', color='orange')
plt.title('Total Flights per Year')
plt.xlabel('Year')
plt.ylabel('Total Flights (in Million)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```
